ENJERCICIO 1

```python
alumnos = {}
while True:
    try:
        cant = int(input("Ingrese la cantidad de alumnos: "))
        if cant > 0:
            break
        else:
            print("El valor debe ser positivo")
    except ValueError:
        print("Valor incorrecto")
        
for i in range(cant):
    nom = input("Nombre del alumno: ")
    if nom.isalpha() == True:
        while nom in alumnos:
            print("Alumno ya existente")
            nom = input("Nombre del alumno: ")
        notas = []
        nota = int(input("Ingresar notas (negativo para fin): "))
        while nota > 0:
            notas.append(nota)
            nota = int(input("Ingresar notas (negativo para fin): "))
        alumnos[nom]=notas[:]
    else: 
        print("El nombre deben ser caracteres")

for nom,notas in alumnos.items():
    promedio = sum(notas) / len(notas)
    print (f"{nom} ha optenido un promedio de {promedio}" )
    
```
